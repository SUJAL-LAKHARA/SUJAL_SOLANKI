TWO sorted median of array


```python
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        merged =(sorted(nums1 + nums2))
        n = len(merged)
        mid = n // 2
        if n % 2:
            return merged[mid]
        return float(merged[mid] + merged[mid - 1]) / 2
```
